```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('hepatitis_cleaned2.csv')
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
df.head(9)
```

```python
#assigning the features and outputs
x = np.array(df.loc[:,['AGE','SEX','STEROID','ANTIVIRAL','FATIGUE','MALAISE','ANOREXIA','LIVER_BIG','LIVER_FIRM','SPLEEN_PALPABLE','SPIDERS','ASCITES','VARICES','BILIRUBIN','ALK_PHOSPHATE','SGOT','ALBUMIN','PROTIME','HISTOLOGY']])
y = np.array(df.loc[:,['DIE_LIVE']])
#print(y)
#x.shape
```

```python
#Assigning the weight values 
#for the input neurons
w1 = np.random.rand(19)
w2 = np.random.rand(19)
w3 = np.random.rand(19)

#Learning rate of the model
lr = 0.001

#for the output neuron 
w4 = np.random.rand(3)

#splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.28, random_state=3)

[r,c] = X_train.shape
e=0
#print(w1)
epoch = 1000
while(e<epoch):
    for i in range(r):
        s1=0
        s2=0
        s3=0
        for j in range(c):
                s1+=w1[j]*x[i,j]
                s2+=w2[j]*x[i,j]
                s3+=w3[j]*x[i,j]
        out1=sigmoid(s1)
        out2=sigmoid(s2)
        out3=sigmoid(s3)
        s4=0
        #progating to the output layer
        s4+= (out1*w4[0]+out2*w4[1]+out3*w4[2])
        out4=sigmoid(s4)
        #calculating the delta value for the output neuron
        err4=out4*(1-out4)*(y_train[i]-out4)
        #updating the weights of the output layer
        w4[0] = w4[0] + (lr*err4*out1)
        w4[1] = w4[1] + (lr*err4*out2)
        w4[2] = w4[2] + (lr*err4*out3)
        #BackPropagating/Calculating the hidden layer delta values
        err1=out1*(1-out1)*(w4[0]*err4)
        err2=out2*(1-out2)*(w4[1]*err4)
        err3=out3*(1-out3)*(w4[2]*err4)
        #Updating the weights of the input layer
        for j in range(c):
            w1[j] = w1[j] + (lr*err1)
            w2[j] = w2[j] + (lr*err2)
            w3[j] = w3[j] + (lr*err3)
    e+=1

#print(w1)
```

```python
# Testing the model 
[r,c] = X_test.shape
y_pred = []
for i in range(r):
    s1=0
    s2=0
    s3=0
    for j in range(c):
            s1+=w1[j]*X_test[i,j]
            s2+=w2[j]*X_test[i,j]
            s3+=w3[j]*X_test[i,j]
    out1=sigmoid(s1)
    out2=sigmoid(s2)
    out3=sigmoid(s3)
    s4=0
    s4+=(out1*w4[0]+out2*w4[1]+out3*w4[2])
    out4=sigmoid(s4)
    y_pred.append(out4)

print('Y_test\n')
print(y_test)
print('y_pred\n')
print(y_pred)
```
